## Performing Visualization(EDA)

```python
# Sort the dataframe by "Population" in ascending order
df_s = df_n.sort_values(by="Population", ascending=True)

# figure size
plt.figure(figsize=(20, 9))

# Plot the bar graph
plt.bar(df_s["Name"], df_s["Population"])

# Rotate the x-axis labels for better visibility
plt.xticks(rotation=90)

# Set labels and title
plt.xlabel("State Name")
plt.ylabel("Population")
plt.title("Population by State")

# Display the plot
plt.show()
```

```python
# Sort the dataframe by "Literacy Rate" in ascending order
df_s = df_n.sort_values(by="Literacy Rate", ascending=True)

# figure size
plt.figure(figsize=(20, 9))

# Plot the bar graph
plt.bar(df_s["Name"], df_s["Literacy Rate"])

# Rotate the x-axis labels for better visibility
plt.xticks(rotation=90)

# Set labels and title
plt.xlabel("State Name")
plt.ylabel("Literacy Rate")
plt.title("Literacy Rate by State")

# Display the plot
plt.show()
```

```python
# Sort the dataframe by "Area(sq.km)" in ascending order
df_s = df_n.sort_values(by="Area(sq.km)", ascending=True)

# figure size
plt.figure(figsize=(20, 9))

# Plot the bar graph
plt.bar(df_s["Name"], df_s["Area(sq.km)"])

# Rotate the x-axis labels for better visibility
plt.xticks(rotation=90)

# Set labels and title
plt.xlabel("State Name")
plt.ylabel("Area(sq.km)")
plt.title("Area(sq.km) of State")

# Display the plot
plt.show()
```

```python
# Now we plot different graphs for each columns.
def visual_h(df_n,i):
    plt.figure(figsize=(25,10),facecolor='red')
    plt.xticks(rotation=80)
    sns.histplot(x=i,data=df_n,kde=True)
```

```python
visual_h(df_n,'Population')
```

```python
visual_h(df_n,'Literacy Rate')
```

```python
visual_h(df_n,'Area(sq.km)')
```

```python
def visual_c(df_n,i):
    plt.figure(figsize=(25,10),facecolor='green')
    plt.xticks(rotation=75)
    sns.countplot(x=i,data=df_n)
```

```python
visual_c(df_n,'Population')
```

```python
visual_c(df_n,'Literacy Rate')
```

```python
visual_c(df_n,'Area(sq.km)')
```

```python
# Remove commas from the "Population" column as we know this column is in string
df_n['Population'] = df_n['Population'].str.replace(',', '')

# Convert the "Population" column from string to numeric (float or integer)
df_n['Population'] = pd.to_numeric(df_n['Population'])

# Calculate the population sum for each category
population_sum = df_n['Population'].sum()

# Calculate the population proportion for each category
pop_p = df_n['Population'] / population_sum
```

```python
plt.figure(figsize=(15, 15))
plt.pie(pop_p, labels=df_n['Name'], autopct='%1.1f%%',rotatelabels=65)
plt.title("Population Distribution")
plt.show()
```

```python
# Remove the percentage symbols (%) from the "Literacy Rate" column
df_n['Literacy Rate'] = df_n['Literacy Rate'].str.rstrip('%').astype('float')
```

```python
plt.figure(figsize=(15, 15))
plt.pie(df_n['Literacy Rate'], labels=df_n['Name'], autopct='%1.1f%%',rotatelabels=65)
plt.title("Literacy Rate Distribution")
plt.show()
```

```python
df_n['Area(sq.km)'] = df_n['Area(sq.km)'].str.replace(',', '')
```

```python
plt.figure(figsize=(15, 15))
plt.pie(df_n['Area(sq.km)'], labels=df_n['Name'], autopct='%1.1f%%',rotatelabels=65)
plt.title("Area Distribution")
plt.show()
```
