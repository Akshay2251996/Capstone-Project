#### 3.2	Write SQL query to find the highest number of rapes & Kidnappings that happened in which state, District, and year

```python
# For max rape
cursor.execute("SELECT state_name, district_name, year FROM womencrime_2001_2012 WHERE rape = (SELECT MAX(rape) FROM womencrime_2001_2012)")

# Fetch the result
result = cursor.fetchone()

# Print the result if it exists
if result:
    print("The following list of years in which the most rapes occurred in the state in his district:")
    print("*"*30)    
    print("State: ", result[0])
    print("District: ", result[1])
    print("Year: ", result[2])
else:
    print("No records found.")
```

```python
# For max kidnapping_and_abduction
cursor.execute("SELECT state_name, district_name, year FROM womencrime_2001_2012 WHERE kidnapping_and_abduction = (SELECT MAX(kidnapping_and_abduction) FROM womencrime_2001_2012)")

# Fetch the result
result = cursor.fetchone()

# Print the result if it exists
if result:
    print("The following list of years in which the most kidnapping and abductions occurred in the state in his district:")
    print("*"*30)
    print("State: ", result[0])
    print("District: ", result[1])
    print("Year: ", result[2])
else:
    print("No records found.")
```

## (c) Crime Comitted IPC 2001-2012 (3.8.1-3.8.4)

#### 3.8.1	Insert records from 01_District_wise_crimes_committed_IPC_2001_2012.csv into a new table.

```python
# Creating database and declaring cursor
db=sqlite3.connect("IPCCrimes_database.db")
cursor=db.cursor()
```

```python
cursor.execute("CREATE TABLE crimedata_ipc(state_ut VARCHAR(100),district VARCHAR(100),year INT,murder INT,attempt_to_murder INT,rape INT)")
```

```python
import csv

# CSV file path
csv_file_path = '01_District_wise_crimes_committed_IPC_2001_2012.csv'

# Open the CSV file
with open(csv_file_path, 'r') as file:
    # Create a CSV reader
    csv_reader = csv.reader(file)
    
    # Skip the header row
    next(csv_reader)

    # Counter for the number of records inserted
    records_inserted = 0

    # Iterate over the rows and insert data into the table
    for row in csv_reader:
        state_ut = row[0]
        district = row[1]
        year = int(row[2])
        murder = int(row[3])
        attempt_to_murder = int(row[4])
        rape = int(row[6])

        # Insert the data into the table
        insert_query = "INSERT INTO crimedata_ipc(state_ut, district, year, murder, attempt_to_murder, rape) VALUES (?, ?, ?, ?, ?, ?)"
        data = (state_ut, district, year, murder, attempt_to_murder, rape)
        cursor.execute(insert_query, data)

        # Increment the counter
        records_inserted += 1

# Print the number of records inserted
print("Number of records inserted:", records_inserted)

# Commit the changes and close the connection
db.commit()
```

```python
# this is to fetch the record from the table.
results=cursor.execute("SELECT * FROM crimedata_ipc")      #Checking for records
for row in results:
    print(row)
```

#### 3.8.2   Write SQL query to find which District in each state/ut has the highest number of murders yearwise. Your output should show STATE/UT, YEAR, DISTRICT, and MURDERS.

```python
cursor.execute("SELECT state_ut, year, district, murder FROM ( SELECT state_ut, year, district, murder,ROW_NUMBER() OVER (PARTITION BY state_ut, year ORDER BY murder DESC) AS rn FROM crimedata_ipc) AS ranked WHERE rn = 1")
# Fetch all the rows returned by the query
rows = cursor.fetchall()

# Display the results
for row in rows:
    print(row)
```

#### 3.8.3	Store the above data (the result of 3.2) in DataFrame and analyze districts that appear 3 or more than 3 years and print the corresponding state/ut, district, murders, and year in descending order.

```python
#  the query is
query = '''
SELECT state_ut, year, district, murder
FROM (
    SELECT state_ut, year, district, murder,
           ROW_NUMBER() OVER (PARTITION BY state_ut, year ORDER BY murder DESC) AS rn
    FROM crimedata_ipc
) AS ranked
WHERE rn = 1
'''

cursor.execute(query)

# Fetch all the rows returned by the query
rows = cursor.fetchall()
```

```python
import pandas as pd
# Creating a DataFrame from the fetched rows
df = pd.DataFrame(rows, columns=['state_ut', 'year', 'district', 'murder'])
```

```python
# Counting the number of years each district appears in
district_counts = df['district'].value_counts()
```

```python
# Filter districts that appear in 3 or more years
filtered_df = df[df['district'].isin(district_counts.index[district_counts >= 3])]
```

```python
# Sorting the filtered DataFrame by year in descending order
sorted_df = filtered_df.sort_values(by='year', ascending=False)
```

```python
# Print the corresponding state/UT, district, murders, and year
print(sorted_df[['state_ut', 'district', 'murder', 'year']])
```

#### 3.8.4	Use appropriate graphs to show your data (the result of 3.8.3)

```python
import matplotlib.pyplot as plt

# Counting the number of occurrences of each district
district_counts = df['district'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(20,8))
district_counts.plot(kind='bar')
plt.xlabel('District')
plt.ylabel('Count')
plt.title('Number of Occurrences of Districts')
plt.show()
```

```python
# Grouping the data by year and calculating the total number of murders per year
murders_per_year = df.groupby('year')['murder'].sum()

# Plot the line chart
plt.figure(figsize=(20,8))
murders_per_year.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Total Murders')
plt.title('Total Murders per Year')
plt.show()
```

```python
# Calculating the percentage of murders by district
district_percentages = df['district'].value_counts(normalize=True) * 100

# Plotting the pie chart
plt.figure(figsize=(18,18))
district_percentages.plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')
plt.title('Percentage of Murders by District')
plt.show()
```

```python
# Plot a histogram of murder counts
plt.figure(figsize=(20,8))
plt.hist(df['murder'], bins=10)
plt.xlabel('Murder Count')
plt.ylabel('Frequency')
plt.title('Distribution of Murder Counts')
plt.show()
```

```python
# Plot a scatter plot of Year vs. Murder count
plt.figure(figsize=(20,8))
plt.scatter(df['year'], df['murder'])
plt.xlabel('Year')
plt.ylabel('Murder Count')
plt.title('Murder Count by Year')
plt.show()
```

```python
# a box plot of murder counts
plt.boxplot(df['murder'])
plt.ylabel('Murder Count')
plt.title('Distribution of Murder Counts')
plt.show()
```
